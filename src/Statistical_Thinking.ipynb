```python
import pandas as pd
import seaborn as sns
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

mgd = pd.read_csv("../data/MDG_Export_20191227.csv")
mgd.head(11)
```

```python
len(mgd['Country'].unique())
mgd = mgd[mgd != 0]

mgd.head(11)
```

```python
mgd = mgd.dropna(how='all',axis ='columns')
mgd = mgd.dropna(subset = ["1990"])
mgd.head()
```

```python

co2 = mgd.loc[mgd['Series'] == "Carbon dioxide emissions (CO2), thousand metric tons of CO2 (CDIAC)"]

co2.sort_values(by='1990').head()
```

```python
co2.sort_values(by='1990').tail()
```

```python
mgd['1990'] = pd.to_numeric(mgd['1990'])

mgd.describe()
```

```python
#this the median
print("this the median:", np.median(co2['1990']))
#this is the mean
print("this the min:" ,np.mean(co2['1990']))
#this is the min
print("this the min:" ,np.min(co2['1990']))
#this is the max
print("this the max:" ,np.max(co2['1990']))
#this STD
print("this is the standard deviation:", np.std(co2['1990']))

print("this is the interquatile range:" ,sp.stats.iqr(co2['1990']))

print("this is the starndard error:" ,co2['1990'].sem())
```
